Desarrolla un script que liste en pantalla (por consola) los elementos HTML que posee y los valores de cada uno de sus atributos.


```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

A continuación, le indicaremos el link proporcionado que se abra en Chrome

```python
driver = webdriver.Chrome() 
url = 'https://jumafernandez.github.io/soco-web_scraping/data/encuentro-02/formulario-tp.html'

driver.get(url)
```

Finalmente, obtenemos el resultado de los elementos contenidos en el url

```python
print("Punto A: Listando elementos HTML y sus atributos y valores\n")
elements = driver.find_elements(By.XPATH, "//*")
for element in elements:
    print("Elemento:", element.tag_name)
    attributes = element.get_attribute("outerHTML")
    print("  Atributos y valores:", attributes)
    print()
```

<span style="font-size: 15px; font-weight: bold;">PUNTO 1. B</span><br><br>
Desarrolla un script que emule la interacción con todos los elementos HTML que posea para luego presionar el botón ‘Ingresar’.


Primero, presiono el botón 'Ingresar'

```python
wait = WebDriverWait(driver, 10)
wait.until(EC.presence_of_element_located((By.ID, "formulario")))
```

Luego, debo ingresar el mail

```python
email = driver.find_element(By.ID, "email")
email.send_keys("daliggg.dg@gmail.com")
```

Encuentro la casilla contraseña e ingreso una 

```python
contrasena = driver.find_element(By.ID, "passwd")
contrasena.send_keys("dalila1234")
```

Selecciono mi país en el desplegable

```python
pais = driver.find_element(By.ID, "pais")
argentina = pais.find_element(By.XPATH, "//option[contains(text(), 'Argentina')]")
argentina.click()
```

Clickeo "Recordar Usuario"

```python
recordar_usuario = driver.find_element(By.ID, "recordar")
recordar_usuario.click()
```

Finalmente, presiono "Ingresar"

```python
ingresar_button = driver.find_element(By.CSS_SELECTOR, "button[type='submit']")
ingresar_button.click()
```
