```python
from bs4 import BeautifulSoup
import requests

# Indico el link
url = "https://raw.githubusercontent.com/jumafernandez/soco-web_scraping/main/data/encuentro-01/example-01.html"
response = requests.get(url)

# Verifico si la solicitud fue exitosa (código 200)
if response.status_code == 200:
    # Creo un objeto BeautifulSoup a partir del contenido HTML
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Encuentro todos los elementos <td> que tengan la clase "marca"
    td_elements = soup.find_all('td', class_='marca')
    
    # Itero sobre los elementos encontrados
    for td in td_elements:
        # Verifico si el texto del elemento termina con la letra "o"
        if td.get_text().strip().endswith('o'):
            # Muestro en pantalla el texto del elemento
            print(td.get_text().strip())
else:
    print("Error", response.status_code)
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

# Indico el link
url = "https://raw.githubusercontent.com/jumafernandez/soco-web_scraping/main/data/encuentro-01/example-01.html"
response = requests.get(url)

# Verifico si la solicitud fue exitosa (código 200)
if response.status_code == 200:
    # Creo un objeto BeautifulSoup a partir del contenido HTML
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Encuentro la tabla en el HTML
    table = soup.find('table')
    
    # Extraigo los datos de la tabla y los convierto en una lista de listas
    rows = []
    for row in table.find_all('tr'):
        rows.append([cell.get_text(strip=True) for cell in row.find_all(['th', 'td'])])

    # Creo un DataFrame de Pandas a partir de la lista de listas
    df = pd.DataFrame(rows[1:], columns=rows[0])
    
    # Selecciono solo las columnas "tienda" y "sitio web"
    df = df[['Nombre', 'Enlace']]
    
    # Guardo el DataFrame en un archivo Excel
    df.to_excel("tabla_tradicionales.xlsx", index=False)
    
    print("Tabla guardada correctamente en el archivo 'tabla_tradicionales.xlsx'")
else:
    print("Error", response.status_code)
```
